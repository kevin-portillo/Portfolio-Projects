```python
import random

class Board():
    #used to create a board with given spaces and to get size
    def __init__(self, spaces=10):
        self.spaces=spaces
    
    def size(self) ->int:
        return self.spaces

class Dice():
    #create dice given numofSides and function to generate a random roll
    def __init__(self, sides=6):
        self.sides=sides
    
    def roll(self) ->int:
        roll=random.randint(1,self.sides)
        #print("Roll:", roll)
        return roll

class Player():
    #create Player using given name and see position of player
    def __init__(self, name):
        self.position=0
        self.name=name
    
    def getName(self):
        return self.name
    
    def getPosition(self):
        return self.position
    
def startGame():
    #create global variables(board, dice1, and global list of Player objects)
    global board
    board=Board()
    
    global dice1
    dice1=Dice()
    
    numPlayers=int(input("Num of players: "))
    global players 
    players=list()
    
    for i in range(numPlayers):
        name=input(f"What is player {i+1}'s name?: ")
        players.append(Player(name))
    print("**********************")
    print("Players: ")
    
    for player in players:
        print(player.getName())
    print("***********************")

def continueGame(boardSize, playerList, dicePiece):
    
    gameActive=True
    turn=1
    while gameActive:
        #Print board size and positions of all players
        print()
        print("Board size: ", boardSize)
        print("Positions: ")
        for player in playerList:
            print(player.getName(),":",player.getPosition())
        print("--------")
        
        #print the turn and go through each player having them roll and checking to see if they won
        print("Turn: ", turn)
        for player in playerList:
            input(f"{player.getName()} press enter to roll")
            spaces=dicePiece.roll()
            print(player.getName(), "'s roll: " ,spaces)
            player.position+=spaces
        
            if player.position>=boardSize:
                print(player.getName(), "has won!!!!!!!")
                gameActive=False
                #break
        turn+=1
        print("******************")
        
def endGame(playerList):
    #print final positions
    print()
    print("*****FINAL RESULTS*****")
    for player in playerList:
        print(player.getName(),":", player.getPosition() )

startGame()
continueGame(board.size(),players,dice1)
endGame(players)
```
