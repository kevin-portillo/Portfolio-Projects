## Hangman

```python
import random
#list with 7 hangman pics
hangman_pics=["""
 +---+
      |
      |
      |
     ===""","""
 +---+
 o    |
      |
      |
     ===""","""
 +---+
 o    |
 |    |
      |
     ===""","""
 +---+
 o    |
/|    |
      |
     ===""","""
 +---+
 o    |
/|\   |
      |
     ===""","""
 +---+
 o    |
/|\   |
/     |
     ===""","""
 +---+
 o    |
/|\   |
/ \   |
     ==="""]

#create list of (61) words after splitting on space
# \(backslash) is used to continue code on next line 
words="ant baboon badger bat bear beaver camel cat clam cobra cougar coyote crow deer dog donkey duck eagle ferret fox\
frog goat goose hawk lion lizard llama mole monkey moose mouse mule newt otter owl panda parrot pigeon python rabbit\
ram rat raven rhino salmon seal shark sheep skunk sloth snake spider stork swan tiger toad trout turkey turtle weasel\
whale wolf wombat zebra".split()

#get random word from the list that was passed
def getRandomWord(wordList):
    #get a random number between 0 and listLength-1(inclusive)
    wordIndex=random.randint(0,len(wordList)-1)
    
    return wordList[wordIndex]

#display the current hangman and current missed/correct letters
def displayHangman(missedLetters,correctLetters,secretWord):
    print(hangman_pics[len(missedLetters)])
    print()
    
    #show incorrect guesses
    print("Missed letters:")
    for letter in missedLetters:
        print(letter, end=" ")
    print()
    
    #create string of dashes that is the same length as the secret word 
    blanks="-"*len(secretWord)
    
    #replace dashes in "blanks" string with correct guesses
    for i in range(len(secretWord)):
        if secretWord[i] in correctLetters:
            blanks=blanks[:i]+secretWord[i]+blanks[i+1:]
    
    #print the secret word with correct guesses revealed 
    for letter in blanks:
        print(letter, end=" ")
    
    print()
    print("********************")
    print("********************")


#return letter the player entered and make sure it is 1 letter only
def getGuess(alreadyGuessed):
    while True:
        guess=input("Guess a letter: ")
        guess=guess.lower()
        #check if input has more than one char
        if len(guess)!=1:
            print("Please enter only 1 letter.")
        
        #check if input was guesed previously
        elif guess in alreadyGuessed:
            print("That letter was already guessed. Try another.")
        
        #check if guess is a non-letter character
        elif guess not in "abcdefghijklmnopqrstuvwxyz":
            print("Please enter a LETTER ONLY.")
        
        #if the guess is 1 char long, is a letter, and has not been guessed before, then return the guess
        else:
            return guess

#see if player want to play again
def playAgain():
    playing=input("Do you want to play again? (Y)es or (N)o ")
    #return the boolean after making the input lower case and seeing if it starts with a 'y'
    return playing.lower().startswith("y")


print("HANGMAN")
missed=[]
correct=[]
secret=getRandomWord(words)
gameFinished=False

while True:
    displayHangman(missed,correct,secret)
    #create a list that contains both missed and correct letters guessed
    allGuesses=missed+correct
    
    #let player enter a letter
    guess=getGuess(allGuesses)
    
    #see if guess is in the secret word
    if guess in secret:
        correct.append(guess)
    
        #check if player has won
        foundAllLetters=True
        for i in range(len(secret)):
            if secret[i] not in correct:
                foundAllLetters=False
                break
    
        if foundAllLetters:
            print(f"Yes!!! You won!!! The secret word was {secret}.")
            gameFinished=True
    
    # guess was not in the secret word then do the following    
    else:
        missed.append(guess)
        
        #check if player guessed too many times
        if len(missed) == len(hangman_pics)-1:
            displayHangman(missed,correct, secret)
            print(f"You ran out of guesses! \nAfter {len(missed)} missed guesses and {len(correct)} correct guesses, the word was {secret}.")
            gameFinished=True
            
    #ask if they want to play again if the game is finished
    if gameFinished:
        if playAgain():
            #reset the following to initial values
            missed=[]
            correct=[]
            secret=getRandomWord(words)
            gameFinished=False
            
        else:
            #exit while loop
            break
```
